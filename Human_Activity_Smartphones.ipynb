## Create column target for the prediction modeling

```python
#dictionnary for the column target. 
#In activity : WALKING, WALKINGUPSTAIRS, WALKINGDOWNSTAIRS (1)
#Not in activity : SITTING, STANDING, LAYING (0)
target = {"LAYING": 0, "STANDING": 0, "WALKING": 1, "SITTING": 0, "WALKING_UPSTAIRS": 1, "WALKING_DOWNSTAIRS": 1}
```

## PCA with StandardScaler

PCA is used to reduce the superfluous complexity of the dataset (reduce the dimensions, reducing number of features) in order to speed up the learning of the machine.

```python
#import library
from sklearn.decomposition import PCA
```

```python
modelpca=PCA(n_components=105)#at least 2 even 1 is enough here
Xpca_reduced=modelpca.fit_transform(X_stdscl)
```

```python
Xpca_reduced
np.cumsum(modelpca.explained_variance_ratio_)
```

```python
#reduce the size of the dataset while maintaining 95 to 99% of the variance of the dataset data
plt.plot(np.cumsum(modelpca.explained_variance_ratio_))
```

We see that we manage to keep 95% of the variance of the data from n_components = 105 and 99% of the variance from n_components = 180

## Import libraries from sklearn

## Split into train/validation datasets - features from PCA and Robuscaller

```python
X_train, X_val,y_train,y_val = train_test_split(Xpca_reduced_R, \
                                                  df_trm.Target, \
                                                  test_size=0.2, \#20% in the validation set and 80% in train set
                                                  random_state=42)#test without stratify=df.target because the target is well balanced yet
```

### Decision Tree

```python
model1=DecisionTreeClassifier()

model1.fit(X_train, y_train)
y_pred1=model1.predict(X_val)

conf1=confusion_matrix(y_val, y_pred1)#Compute confusion matrix to evaluate the accuracy of a classification.

acc1=accuracy_score(y_val, y_pred1)
rec1=recall_score(y_val, y_pred1)
pre1=precision_score(y_val, y_pred1)
f11=f1_score(y_val, y_pred1)
display(conf1)

print('Accuracy', acc1)#% of True positive
print('Recall', rec1)#min False Negative
print('Precision', pre1)#min False Positive
print('F1', f11)# 2*(Precision*Recall/Precision+Recall)
```

## import library

## Cross validation

```python
model = KNeighborsClassifier()
cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')
```

```python
val_score = []
for k in range(1, 50):
    score = cross_val_score(KNeighborsClassifier(k), X_train, y_train, cv=5).mean()
    val_score.append(score)

plt.plot(val_score)
```

## import library

```python
from sklearn.model_selection import validation_curve
```

## Validation curve

```python
model = KNeighborsClassifier()
k = np.arange(1, 50)

train_score, val_score = validation_curve(model, X_train, y_train,
                                          'n_neighbors', k, cv=5)

plt.plot(k, val_score.mean(axis=1), label='validation')
plt.plot(k, train_score.mean(axis=1), label='train')

plt.ylabel('score')
plt.xlabel('n_neighbors')
plt.legend()
```
